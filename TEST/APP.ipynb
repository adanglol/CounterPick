```python
# FUNCTIONS TO HANDLE RIOT API

def get_puuid(gameName: str= None,tagLine: str=None,region: str=None,api_key: str=None):
    link = f'https://{region}.api.riotgames.com/riot/account/v1/accounts/by-riot-id/{gameName}/{tagLine}?api_key={api_key}'
    response = requests.get(link)
    return response.json()["puuid"]

def get_match_ids(puuid: str=None,region: str=None,mode: str=None,count: int=20,api_key: str=None):
    link = f'https://{region}.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?queue={queue_id[mode]}&start=0&count={count}&api_key={api_key}'
    response = requests.get(link)
    return response.json()

def get_match_by_id(match_id: str=None, region: str=None, api_key : str=None):
    link = f'https://{region}.api.riotgames.com/lol/match/v5/matches/{match_id}?api_key={api_key}'
    response = requests.get(link)
    return response.json()

def get_match_timeline(match_id: str=None, region: str=None, api_key : str=None):
    link = f'https://{region}.api.riotgames.com/lol/match/v5/matches/{match_id}/timeline?api_key={api_key}'
    response = requests.get(link)
    return response.json()

# IMPLEMENT GETTING RANKED DATA AND COLLECT THEM WITH GOOGLE SHEETS MAYBE USE DB AFTER
```
